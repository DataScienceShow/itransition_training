```python
import hashlib
import os

# 🔹 YOUR INPUTS
FOLDER_PATH = r"C:\Users\Asus\Ziyodullo\LearningIT\LearnDataEngineeering\ItransitionTraining\SourceFiles\Task2\task2"
EMAIL = "ziyodullofficial@gmail.com".lower()


# 🔹 SHA3-256 helper
def sha3_256_bytes(data: bytes) -> str:
    return hashlib.sha3_256(data).hexdigest()


# 🔹 Compute hash of one file
def compute_file_hash(filepath):
    with open(filepath, "rb") as f:
        return sha3_256_bytes(f.read())


# 🔹 Sorting key: product of (hex_digit + 1)
def hash_sort_key(hash_str):
    product = 1
    for ch in hash_str:
        product *= (int(ch, 16) + 1)
    return product


# 🔹 Collect hashes
hashes = []

for root, _, files in os.walk(FOLDER_PATH):
    for file in files:
        if file.endswith(".data"):  # only .data files
            path = os.path.join(root, file)
            file_hash = compute_file_hash(path)
            hashes.append(file_hash)

# 🔹 Sort hashes
hashes_sorted = sorted(hashes, key=hash_sort_key)

# 🔹 Join hashes
joined_hashes = "".join(hashes_sorted)

# 🔹 Append email
final_string = joined_hashes + EMAIL

# 🔹 Final SHA3-256
final_hash = sha3_256_bytes(final_string.encode())

print("Final SHA3-256:", final_hash)
```

```python
import hashlib
import os
from pathlib import Path
from functools import reduce
from operator import mul

# Configuration
FOLDER_PATH = r"C:\Users\Asus\Ziyodullo\LearningIT\LearnDataEngineeering\ItransitionTraining\SourceFiles\Task2\task2"
EMAIL = "ziyodullofficial@gmail.com"  # Fixed typo: gamil -> gmail

def calculate_sha3_256(file_path):
    """Calculate SHA3-256 hash for a binary file"""
    sha3 = hashlib.sha3_256()
    with open(file_path, 'rb') as f:
        sha3.update(f.read())
    return sha3.hexdigest()

def calculate_sorting_key(hash_hex):
    """Calculate product of (hex_digit + 1) for sorting"""
    product = 1
    for char in hash_hex:
        digit = int(char, 16)  # Convert hex char to int (0-15)
        product *= (digit + 1)
    return product

def main():
    # Step 1: Get all .dat files and calculate their hashes
    print(f"Reading files from: {FOLDER_PATH}")
    file_hashes = []
    
    dat_files = sorted(Path(FOLDER_PATH).glob("*.dat"))
    
    if len(dat_files) != 256:
        print(f"WARNING: Found {len(dat_files)} files, expected 256!")
    
    for file_path in dat_files:
        file_hash = calculate_sha3_256(file_path)
        file_hashes.append(file_hash)
        print(f"Processed: {file_path.name} -> {file_hash}")
    
    print(f"\nTotal files processed: {len(file_hashes)}")
    
    # Step 2: Sort hashes by product of (hex_digits + 1)
    sorted_hashes = sorted(file_hashes, key=calculate_sorting_key)
    
    # Debug: Show first few sorted hashes with their keys
    print("\nFirst 5 sorted hashes with their sorting keys:")
    for i, h in enumerate(sorted_hashes[:5]):
        key = calculate_sorting_key(h)
        print(f"{i+1}. Key: {key}, Hash: {h}")
    
    # Step 3: Join sorted hashes without separator
    concatenated_hashes = ''.join(sorted_hashes)
    
    # Step 4: Concatenate with email in lowercase
    final_string = concatenated_hashes + EMAIL.lower()
    
    print(f"\nConcatenated string length: {len(concatenated_hashes)}")
    print(f"Email appended: {EMAIL.lower()}")
    print(f"Final string length: {len(final_string)}")
    
    # Step 5: Calculate SHA3-256 of the final string
    final_hash = hashlib.sha3_256(final_string.encode('utf-8')).hexdigest()
    
    print(f"\n{'='*70}")
    print(f"FINAL RESULT:")
    print(f"{'='*70}")
    print(f"Hash: {final_hash}")
    print(f"\nSubmit command:")
    print(f"!task2 {EMAIL.lower()} {final_hash}")
    print(f"{'='*70}")
    
    return final_hash

if __name__ == "__main__":
    result = main()
```

```python
import os
import hashlib

# 🔧 SET YOUR PATH
FOLDER_PATH = r"C:\Users\Asus\Ziyodullo\LearningIT\LearnDataEngineeering\ItransitionTraining\SourceFiles\Task2\task2"
EMAIL = "ziyodullofficial@gamil.com".lower()

def sha3_256_hex(data: bytes) -> str:
    return hashlib.sha3_256(data).hexdigest()

def sorting_key(h: str) -> int:
    prod = 1
    for c in h:
        prod *= (int(c, 16) + 1)
    return prod

# 📂 collect ALL .data files (recursive)
files = []
for root, _, filenames in os.walk(FOLDER_PATH):
    for fname in filenames:
        if fname.endswith(".data"):
            files.append(os.path.join(root, fname))

# ✅ MUST be exactly 256 files
if len(files) != 256:
    raise ValueError(f"256 ta .data file bo‘lishi kerak, hozir: {len(files)}")

# 🔐 hash each file
hashes = []
for path in files:
    with open(path, "rb") as f:
        data = f.read()
    hashes.append(sha3_256_hex(data))

# 🔢 sort by custom key
hashes.sort(key=sorting_key)

# 🔗 join hashes (no separator)
joined = "".join(hashes)

# ➕ append email
final_input = (joined + EMAIL).encode("ascii")

# 🔥 final hash
final_hash = sha3_256_hex(final_input)

print(final_hash)
print(f"!task2 {EMAIL} {final_hash}")
```
